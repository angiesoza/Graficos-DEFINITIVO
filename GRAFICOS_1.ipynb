```
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2, 9, 0.05)
y = np.sin(x*np.pi)

fig, ax = plt.subplots(figsize=(3,2), constrained_layout=True)
ax.plot(x, y)
ax.set_xlabel('t [s]')

ax.set_ylabel('S [V]')
ax.set_title('Sine wave')
fig.set_facecolor('lightsteelblue')
```

```
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(2, 9, 0.05)
y = np.sin(x * np.pi)

fig, ax = plt.subplots(figsize=(15, 6), constrained_layout=True)
ax.plot(x, y)
ax.set_xlabel('t [s]')
ax.set_ylabel('S [V]')
ax.set_title('Numero de habitaciones por habitante')
ax.plot(x, y, color='black')

#codigo editado del ejemplo, buscar como cambiar las formitas
fig.set_facecolor('pink')

plt.show()
```
